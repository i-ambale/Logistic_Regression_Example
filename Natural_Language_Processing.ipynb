### 1. Tokenization

Tokenizing is a form of group grouping things. 
We can have two forms;
1. Word tokenizers (separates by words)
2. Sentence tokenizers (separates by sentence)

lexicon are words and their meaning.

corpora is a body of text around the same thing. e.g. medical journals, presidential speeches



```python
from nltk.tokenize import sent_tokenize, word_tokenize
example_text = "Hello Mr. Smith, how are you doing today? The weather is great and Python is awesome. The sky is pinkish-blue. You should not eat cardboard."
print(sent_tokenize(example_text))
```

```python
print(word_tokenize(example_text))
```

```python
# a function to tokenize

for i in word_tokenize(example_text):
    print(i)
```

```python
example2_text = "I can't reject you girl, because your P-value is greater than 0.5. You're one in a million. Be mine little woman."

# separate into sentenses

print(sent_tokenize(example2_text))
```

```python
# separate into words
print(word_tokenize(example2_text))
```

```python
# separate into words using a function
for w in word_tokenize(example2_text):
    print(w)
```

### 2. Stop words

Python has inbuilt/predefined English stop words that can be accessed using the stopword function. 

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# show English stopwords in Python
stopwords.words("english")
```

```python
# show Arabic stopwords in Python
stopwords.words("arabic")
```

```python
# show Spanish stopwords in Python
spanish_stpwords = set(stopwords.words("spanish"))
spanish_stpwords
```

```python
example_sentence = "This is an example showing off stop word filtration."

stop_words = set(stopwords.words("english"))

words = word_tokenize(example_sentence)

filtered_sentence = []

for j in words:
    if j not in stop_words:
        filtered_sentence.append(j)

print(filtered_sentence)
```

```python
filtered_sentence = [j for j in words if not j in stop_words]
print(filtered_sentence)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()

examplelist_words = ["python", "pythoner", "pythoning", "pythoned"]

for k in examplelist_words:
()    print(ps.stem(k))
```

```python
new_text = "It is very important to be pythonly while you are pythoning with python. All pythoners have pythoned poorly at least once."
stem_words = word_tokenize(new_text)
for s in stem_words:
    print(ps.stem(s))
```

```python
from nltk.tokenize import TreebankWordTokenizer

tree_text = "It is very important to be pythonly while you are pythoning with python. All pythoners have pythoned poorly at least once."
token_words = TreebankWordTokenizer()
tokens = token_words.tokenize(tree_text)
print(tokens)

```
